```python
# exp 15
# riya latkar
# 25070123092
# aim: Data normalisatin and data type conversion

import pandas as pd
import numpy as np

data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price': [55000, 20000, 30000, 2000, 8000, 45000],
    'Units_Sold': [25, 60, 35, 80, 50, 15],
    'Discount': [10, 5, 15, 0, 10, 20]
}

df = pd.DataFrame(data)
display(df, "\n")
```

```python
## 1. min max normalisaton
#min max normalisation for price
df["Price_MinMax"] = (df["Price"] - df["Price"].min()) / (df["Price"].max() - df["Price"].min())
display(df[['Product', 'Price', 'Price_MinMax']] )

#min max normalisation for discount
df["Discount_MinMax"] = (df["Discount"] - df["Discount"].min()) / (df["Discount"].max() - df["Discount"].min())
display(df[['Product', 'Discount', 'Discount_MinMax']] )

#min max normalisation for units sold
df["Units_Sold_MinMax"] = (df["Units_Sold"] - df["Units_Sold"].min()) / (df["Units_Sold"].max() - df["Units_Sold"].min())
display(df[['Product', 'Units_Sold', 'Units_Sold_MinMax']])
```

```python
# Normalise Multiple Columns
cols = ["Price", "Units_Sold", "Discount"]
df_norm = (df[cols]- df[cols].min()) / (df[cols].max() - df[cols].min())
display(df_norm)
```

```python
## 2. Z-score optimisation
df["Units_Zscore"] = (df["Units_Sold"] - df["Units_Sold"].mean())/ df["Units_Sold"].std()
display(df[['Product', 'Units_Sold', 'Units_Zscore']])
```

```python
## 3. Decimal Scaling
df["Units_Sold_Scaled"] = df["Units_Sold"] / 100
df["Discount_Scaled"]= df["Discount"] / 100
display(df[['Product', 'Units_Sold', 'Units_Sold_Scaled', 'Discount_Scaled']])
```

```python
# Data Normalisation
import pandas as pd
df = pd.read_csv("amazon_products_dataset_Expt-14.csv")
display(df)
```

```python
# Min-Max Normalisation (all)
#min max normalisation for price
df["Price_MinMax"] = (df["Price"] - df["Price"].min()) / (df["Price"].max() - df["Price"].min())
display(df[['Product_Name', 'Price', 'Price_MinMax']] )
print("\n")

#min max normalisation for rating
df["Rating_MinMax"] = (df["Rating"] - df["Rating"].min()) / (df["Rating"].max() - df["Rating"].min())
display(df[['Product_Name', 'Rating', 'Rating_MinMax']] )
print("\n")

#min max normalisation for units sold
df["Units_Sold_MinMax"] = (df["Units_Sold"] - df["Units_Sold"].min()) / (df["Units_Sold"].max() - df["Units_Sold"].min())
display(df[['Product_Name', 'Units_Sold', 'Units_Sold_MinMax']])
print("\n")
```

```python
#Normalising multiple columns
cols = ["Price", "Units_Sold", "Rating"]
df_norm = (df[cols]- df[cols].min()) / (df[cols].max() - df[cols].min())
display(df_norm)
```

```python
# Z-score normalisation
df["Units_Zscore"] = (df["Units_Sold"] - df["Units_Sold"].mean())/ df["Units_Sold"].std()
display(df[['Product_Name', 'Units_Sold', 'Units_Zscore']])
```

```python
# Decimal Scaling (reviews and price)
df["Reviews_Scaled"] = df["Reviews"] / 1000
df["Price_Scaled"]= df["Price"]/100000
display(df[['Product_Name', 'Reviews', 'Reviews_Scaled','Price', 'Price_Scaled']])
```
